### 1.2 使用因子

1. **消費者物價指數CPI:**（公佈時間：每月會公佈一次。）
-  CPI年增率上升，貨幣政策會傾向緊縮，股價下跌，CPI年增率下滑，貨幣政策會傾向寬鬆，股價上漲。
2. **失業率:**（公佈時間：每月會公佈一次。）
-  因為失業率可能會影響投資人對市場的預期，當失業率上升可能代表當今社會經濟不景氣，投資人對市場的信心度也會有所下降，從而減少資金的投入等待時機更好時再投入。但當失業率下降後，投資人對市場的預期也會變高，並增加資金的投入。
3. **半導體機械進口數量:**（公佈時間：每月會公佈一次。）
-  希望可以藉由生產半導體等機械進口數量，來推測未來半導體廠商對於之後半導體產業發展的看法，以及需求進而影響未來出口變化。近年台灣半導體產業是提升優化減少製程時間，由於通訊軟體5G、人工智慧、物聯網等新興科技應用，皆需使用晶片，因此半導體設備進口需求大多為正向的百分比。近年除2017年至2019年左右，受到疫情影響，工廠因隔離政策導致供應鏈問題，當時成長百分比為負之外，半導體需求大多為正向成長的。
4. **生產者價格指數PPI:**（公佈時間：每月會公佈一次。）
-  生產者物價指數是一種先行指標，他可以衡量企業需購買製造商所生產的商品與服務時，每個不同時間點付出的成本為何，以便評估批發價格的變化狀況。 
5. **費城半導體(SOX):**（公佈時間：每秒鐘公布）
-  費城半導體指數由30檔成分股組成，須在美國掛牌上市的公司，其中也包含台灣人所熟知的台積電ADR，還有一些我們常聽到的晶片大廠高通(QUALCOMM)、超微(AMD)、英特爾(Intel)，半導體製造設備與材料大廠艾斯摩爾(ASML)、應用材料(AMAT)等，也因為半導體是許多科技產品的基礎，因此這個產業也被視為全球科技產業的領頭羊。
6. **聯邦基本利率:**（公佈時間：每月公布）
-  升息讓投資及投機的機會成本提高，人們更願意將資金放在保守的定存。升息會讓借錢變困難、資金成本提高，可以抑制投機熱錢。

**自動整合移動平均自迴歸模型（Autoregressive Integrated Moving Average model）**

可以說是整合移動平均自迴歸模型(ARIMA)的進化版。這是一種有助於找到 p、q 和 d 值的最佳組合的模型。由於透過ACF與PACF圖觀察最佳的p、d值時，難免出現錯誤或是無法判斷的情況。因此，使用自動整合移動平均自迴歸模型後，模型便會自動幫我們找出最適當的參數運用在模型中。自動整合移動平均自迴歸模型（Auto ARIMA）參數的組合，是通過使用AIC（Akaike 信息準則）和 BIC（貝葉斯信息準則）來估計。該函數可在使用者設定的邊界值內(max_p, max_q)，搜尋出p、q 和 d 的最佳參數組合。我們就不必人工參考ACF與PACF圖才能決定參數p、d的值。另外Auto ARIMA模型還有一種是可以專門用來預測季節性資料，該函數稱為季節性自回歸綜合移動平均模型(SARIMA)。不過在使用自動整合移動平均自迴歸模型前，還是需要了解整合移動平均自迴歸模型(ARIMA)的模型內容，能夠幫助我們知道自動整合移動平均自迴歸模型產生結果的方式。

**整合移動平均自迴歸模型(ARIMA)**

這是是一個非常經典，用於分析單變量時間序列資預測的方法。整合移動平均自迴歸模型(ARIMA)演算法，結合了自回歸模型（Autoregressive Model），與移動平均模型（Moving Average Models）兩種適用於時間序列分析的模型。以下為ARIMA模型(p,d,q) 三種參數的定義：
1.	p值：表示自回歸模型（Autoregressive Model）中，與過去P期的時間序列資料具有強烈相關性，也稱為滯後階數。
2.	d值：是原始觀測值的差分次數，其功用在於必須將時間數列數據轉為平穩化的資料。
3.	q值：是表示移動平均模型（Moving Average Models）窗口的大小，也稱為移動平均的階數。

(1) p值-自回歸模型（Autoregressive Model）
整合移動平均自迴歸模型參數p值，指的就是自回歸模（Autoregressive Model）。顧名思義，就是在查看自身資料的相關性，模型將自己過去P期時間序列資料做為解釋變數。

(2) d值-差分次數
整合移動平均自迴歸模型(ARIMA)參數ｄ值，指的是原始觀測值的差分次數，其功用在於必須將時間數列數據轉為平穩化的資料。所謂平穩資料必須符合三個特性。第一，數據需具有恆定的平均值。第二，數據需具有恆定方差。第三，數據彼此之間不依賴於時間的自協方差/自相關，簡單來說就是它們是獨立的。有時一階整數差分差數據看起來不是平穩的，可能需要第二次差分數據以獲得平穩序列。因此d值就是代表要多少次差分才可以將資料轉為平穩的資料。

(3) q值-移動平均模型（Moving Average Models）
移動平均模型則是使用過去資料預測的誤差值，在預測下一期的數據，即過去的模型誤差影響未來，過去的誤差可以幫助我們做出預測，並且假設誤差項是相互獨立且常態分佈的。而MA( q ) 指的是階數為q的移動平均模型，階數 q 可以通過觀察時間序列的自相關係數圖（ACF）來決定。

### 2.5.2 使用 SARIMAX 滾動預測

季節性自回歸綜合移動平均模型(SARIMA)它是具有季節性成分Auto ARIMA模型。如同上述介紹，除了自動整合移動平均自迴歸模型舊有（p、d、q）參數外，季節性自回歸綜合移動平均模型(SARIMA)還加入了另一組參數（P、D、Q），以及季節性週期參數m。這將幫助我們在預測金融時間序列數據時，能將季節性因素納入預測與分析。SARIMA模型參數為(p,d,q)(P,D,Q)m，其中(p,d,q)與Auto ARIMA模型相同，可以模型自動訓練找出最佳參數組合。而P、D、Q、m皆為季節參數。P表示季節自回歸的階數。D表示季節性資料需要差分的次數。Q表示季節資料移動平均的階數，m則為季節性週期參數，以每月為單位。

```python
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import r2_score
```

```python
from datetime import datetime
skLR = LinearRegression(fit_intercept=True)
polyReg = LinearRegression( fit_intercept=True)
ridge = Ridge(alpha=1, solver='sag', random_state=42)
lasso = Lasso(alpha=0.01, random_state=42)
SGD = SGDRegressor(max_iter=5000, eta0=0.00001, power_t=0.15, shuffle = False) 
ARD = ARDRegression()
svr_linear = SVR(kernel="linear", C=1000)
# Nonlinear Regressor
regrNN = MLPRegressor(hidden_layer_sizes=(200, 100, 50, 25), activation = 'relu', shuffle = False,
                      alpha=0.0001, random_state=1, max_iter=50000)
svr_nonlinear = SVR(kernel="rbf", C=1000)
```

```python
def train_test_process(listModels, listModelNames, trainData, testData, trainTarget, testTarget):
    print("模型訓練與測試\n")
    modelOLS = sm.OLS(trainTarget, trainData) # The benchmark model
    OLS_fit = modelOLS.fit()
    summary = OLS_fit.summary2()
    print("statsmodels OLS R^2 (Training Data Only) = ", summary.tables[0][0][6], summary.tables[0][1][6])
    prediction = OLS_fit.predict(testData)
    print("statsmodels OLS R^2 (Testing Data Only) = ", r2_score(testTarget, prediction))

    nameList = ['Ordinary Least Squares (OLS)']
    scoreList_Train=[float(summary.tables[0][1][6])]
    scoreList_Test=[r2_score(testTarget, prediction)]
    for aModel, aModelName in zip(listModels, listModelNames):
        process_starttime = datetime.now()
        if aModelName == 'Polynomial Regressor':
            poly = PolynomialFeatures(degree=2, include_bias=False)
            poly_features_train = poly.fit_transform(trainData)
            poly_features_test = poly.fit_transform(testData)
            aModel.fit(poly_features_train, trainTarget)
            y_hat = aModel.predict(poly_features_test)
            print ("\n..... "+aModelName+" 模型訓練花費時間 %s" % (datetime.now() - process_starttime))
            score_r2_train = aModel.score(poly_features_train, trainTarget).round(4) 
            score_r2_test = aModel.score(poly_features_test,  testTarget).round(4)
        else:            
            aModel.fit(trainData, trainTarget)
            y_hat = aModel.predict(testData)
            print ("\n..... "+aModelName+" 模型訓練花費時間 %s" % (datetime.now() - process_starttime))
            score_r2_train = aModel.score(trainData, trainTarget).round(4)
            score_r2_test = aModel.score(testData,  testTarget).round(4)
        nameList.append(aModelName)
        scoreList_Train.append(score_r2_train)
        scoreList_Test.append(score_r2_test)
    resultDF = pd.DataFrame(data = list(zip(nameList, scoreList_Train, scoreList_Test)),
                            columns=['Name of Regressor', 'Training r2', 'Testing r2'])
    resultDF.sort_values("Training r2", axis = 0, ascending = False, inplace = True)
    resultDF.reset_index(drop=True,  inplace = True)
    resultDF['Diff%'] = np.round((resultDF['Training r2']-resultDF['Testing r2'])/resultDF['Training r2'],
                                 4)*100
    return resultDF
```

### 3.4 簡單RNN

是LSTM的基礎，具有記憶的人工神經網路，只運行兩個時間步，儲存輸入層和前一層的訊息並將其處理，傳遞給隱藏層，每個隱藏層裡有不同的激活函數，當要輸出結果時，只會用前一層的訊息來計算，因此有新輸入的加入，原始的輸入值會慢慢被遺忘。

```python
import tensorflow_addons as tfa
tf.random.set_seed(42)
sRNN_model = keras.models.Sequential([
    keras.layers.SimpleRNN(1, input_shape=[None, 1])
])

early_stopping = tf.keras.callbacks.EarlyStopping(monitor="r_square", patience=50, restore_best_weights=True)
opt = tf.keras.optimizers.SGD(learning_rate=0.02, momentum=0.9)
sRNN_model.compile(loss=tf.keras.losses.MeanSquaredError(), optimizer=opt, metrics=tfa.metrics.RSquare())
sRNN_history = sRNN_model.fit(train_ds, validation_data=valid_ds, epochs=100, callbacks=[early_stopping])
```

```python
plot_learning_curves(sRNN_history)
```

### 3.5 深度 RNNs

將深度引入隱藏層使RNN更深，有多個循環隱藏層堆疊起來，並可以將資訊傳遞到後層的多個資訊，最後Dense層生成模型的預測。

```python
tf.random.set_seed(42)
DRNN_model = tf.keras.Sequential([
    tf.keras.layers.SimpleRNN(32, return_sequences=True, input_shape=[None, 1]),
    tf.keras.layers.SimpleRNN(32, return_sequences=True),
    tf.keras.layers.SimpleRNN(32),
    tf.keras.layers.Dense(1)
])
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor="r_square", patience=50, restore_best_weights=True)
opt = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
DRNN_model.compile(loss=tf.keras.losses.MeanSquaredError(), optimizer=opt, metrics=tfa.metrics.RSquare())
DRNN_history = DRNN_model.fit(train_ds, validation_data=valid_ds, epochs=100, callbacks=[early_stopping])
```

```python
plot_learning_curves(DRNN_history)
```

```python
for key in linear_history.history.keys():
    print('linear model', key, linear_history.history[key][-1])
    print('Simple RNN', key, sRNN_history.history[key][-1])
    print('Imporoved Simple RNN', key, univar_history.history[key][-1])
    print('Deep RNN', key, DRNN_history.history[key][-1])
```

**比較linear model loss、Simple RNN、Imporoved Simple RNN loss、Deep RNN r_square**

這邊我們比較了四種模型檢測的結果，可以看到四種模型的loss都落在2.079到3.88之間其中以Deep RNN表現最好，而train loss主要是代表訓練集在模型中的預測結果與真實結果的誤差，因此結果要越小越好。而再r_square皆是負值，因此這四種模型可能都有存在過度擬合的問題。
